```python
import pandas as pd
import numpy as np

ages = pd.Series([18, 20, 19, 22, 21])
print(ages)
ages_np = np.array(ages)
print(ages_np)
```

```python
data = {
    "Name": ["Ali", "Sara", "Omar", "Mona", "Hana"],
    "Age": [18, 20, 19, 22, 21],
    "Math_Score": [85, 70, 90, 60, 75],
    "Gender": ["M", "F", "M", "F", "F"]
}

df = pd.DataFrame(data)
print(df)
print(df.info())
print(df.describe())
```

```python
print(df["Name"])

print(df.loc[0])

print(df.iloc[1])
filtered_df = df[df["Math_Score"] > 70]
print(filtered_df)
```

```python
df.loc[2, "Math_Score"] = np.nan
df.loc[4, "Age"] = np.nan
print(df)
df_filled = df.fillna(df.mean(numeric_only=True))
print(df_filled)
df_dup = pd.concat([df, df.iloc[[0]]])
print(df_dup)

df_no_dup = df_dup.drop_duplicates()
print(df_no_dup)
df_dropped = df.drop(columns=["Age"])
print(df_dropped)
```

```python
def classify_score(score):
    if score >= 60:
        return "Pass"
    else:
        return "Fail"

df["Result"] = df["Math_Score"].apply(classify_score)
print(df)
gender_dummies = pd.get_dummies(df["Gender"])
print(gender_dummies)
```

```python
data2 = {
    "Name": ["Khaled", "Nour"],
    "Age": [23, 19],
    "Math_Score": [88, 65],
    "Gender": ["M", "F"]
}

df2 = pd.DataFrame(data2)

combined_df = pd.concat([df, df2])
print(combined_df)
pivot_table = pd.pivot_table(combined_df, values="Math_Score", index="Gender", aggfunc="mean")
print(pivot_table)
```
