```python
import numpy as np

arr1 = np.array([10, 20, 30, 40, 50])
print(arr1)

arr2 = np.arange(0, 21)
print(arr2)

arr3 = np.linspace(1, 5, 10)
print(arr3)
```

```python
zeros_arr = np.zeros((3, 3))
print(zeros_arr)

ones_arr = np.ones((2, 4))
print(ones_arr)

empty_arr = np.empty((2, 2))
print(empty_arr)

identity_arr = np.identity(3)
print(identity_arr)
```

```python
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

arr2_new = np.arange(0, 20)
reshaped_arr = arr2_new.reshape(4, 5)
print(reshaped_arr)
```

```python
import numpy as np
conditions = [
    arr1 < 10,
    (arr1 >= 10) & (arr1 <= 30),
    arr1 > 30
]
choices = ["Low", "Medium", "High"]
classified = np.select(conditions, choices, default="Unknown")

print("Classified values for arr1:")
print(classified)
```
