```python
# dictionary 
rules = {
    "hello": "Hi! How can I help you?",
    "what is ai": "AI is the simulation of human intelligence.",
    "bye": "Goodbye!",
    "hi": "Hello! Nice to meet you.",
    "who are you": "I am a simple chatbot."
}

def chatbot():
    print("Chatbot is running... (type 'exit' to stop)")

    while True:
        user_input = input("you: ").lower()
        
        if user_input == "exit":
            print("Bot: Bye ")
            break
        found = False
        for question in rules:
            if question in user_input:   # partial matching
                print("Bot:", rules[question])
                found = True
                break
        
        if not found:
            print("Bot: Sorry, I don't understand that yet.")

chatbot()
```
